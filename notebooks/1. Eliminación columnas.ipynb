### Reemplazo NaNs, eliminación y tratmiento de las columnas (basicos)

```python
def preproces(df):
    
    # Limpieza de nombres de columnas
    df.columns = df.columns.str.strip()
    
    # Reemplazo los n.d. y valores vacíos por NaN 
    df.replace({"n.d.": np.nan, "": np.nan, 'n.s.': np.nan}, inplace=True)
    
    # Calculo el porcentaje de NaN por columna
    nan_per = df.isna().mean() * 100
    
    # Filtro las columnas con mas del 80% de NaN
    columnas_mas_80 = nan_per[nan_per > 80].index
    
    # Elimino las columnas con más del 80% de NaN
    df.drop(columns=columnas_mas_80, inplace=True)
    
    return df
```

```python
# Supongamos que 'df' es tu DataFrame original y 'preproces' es tu función
clening_basic_df = preproces(df)

# Guardar el DataFrame en un archivo CSV
clening_basic_df.to_csv('clening_basic_data.csv', index=False)

print("Archivo guardado como 'clening_basic_data.csv'")
```
